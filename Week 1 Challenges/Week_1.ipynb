## Challenge 4:

Write the python code  to find the factorial of a number using a loop. Thefactorial of a number (integer positive) is the sum of multiplication of all the integers smaller than that positive integer.

```python
# selecting a number for the calculation
number = 7

# initiating a res variable that will be manipulated within the loop
res = 1

# looping through all numbers smaller or equal to the selected number and multiplying them
for i in range(1,number + 1):
    res = res*i

print(res)
```

* Write and use a function to convert Fahrenheit to Celsius. Use the input parameter to let the user enter a number in Fahrenheit and then get the correspoingnt value in Celsi.s


```python
# ask user to input temperature
tempF = int(input("Enter temperature in Fahrenheit."))

#define conversion function
def temp_convert(fh):
    tc = (fh - 32)/1.8
    return tc

# print result
print(temp_convert(tempF))
```

## Challenge 7:

Calculate the mean and standard deviation of elements in a NumPy array. The outcome should have something like ``print(f"Mean: {mean_value}, Standard Deviation: {std_deviation}"`


```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])

mean_value = sum(arr1)/len(arr1)
std_deviation = np.std(arr1)

# print results
print(f"Mean: {mean_value}, Standard Deviation: {std_deviation}")
```

## Challenge 8:

You have the following dataframes 'sales_data' and 'movie_ratings'

```python
import pandas as pd

sales_data_dict = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04'],
    'Region': ['North', 'South', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C'],
    'Quantity': [15, 8, 12, 10],
    'Revenue': [1500, 1200, 1800, 900]
}

sales_data = pd.DataFrame(sales_data_dict)


movie_ratings_dict = {
    'Title': ['Movie1', 'Movie2', 'Movie3', 'Movie4'],
    'Genre': ['Action', 'Drama', 'Comedy', 'Thriller'],
    'Year': [2015, 2010, 2018, 2011],
    'Rating': [8.5, 7.8, 9.2, 8.0]
}

movie_ratings = pd.DataFrame(movie_ratings_dict)

print(sales_data)
print(movie_ratings)
```

```python
# first 5 rows of sales_data
sales_data.head()

subset_10_N = sales_data.loc[(sales_data['Quantity'] > 10) & (sales_data['Region'] == "North")]
print(subset_10_N)

# display 1st row and 3 columns with iloc
print(sales_data.iloc[0,0:3])

# total sales for each region
sales_byregion = sales_data.groupby(["Region"])["Revenue"].sum()
print(sales_byregion)

# last 3 rows of movie ratings
print(movie_ratings.tail(3))

# average ratings by genre
rating_genre = movie_ratings.groupby(["Genre"])["Rating"].mean()
print(rating_genre)
```
